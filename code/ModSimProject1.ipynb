```python
filename = r'C:\Users\ator\Desktop\ModSim\ModSimProject1Data.csv'
tables = read_csv(filename, header=0, index_col=0);
```

```python
tables.columns = ['lebt', 'popmp', 'popm', 'popfp', 'popf', 'popt', 'death', 'birth', 'mortm', 'mortf', 'sex']
tables
```

```python
#Before anything, we need to establish some common variables that we'll use throughout the notebook.
popt = tables.popt;
t_0 = get_first_label(popt);
t_ocp = 1979; #when the one-child policy was enacted
t_end = get_last_label(popt);
p_0 = popt[t_0];
p_ocp = popt[t_ocp];
p_end = popt[t_end];
death_rate = tables.death / 1000;
birth_rate = tables.birth / 1000;
growth_rate = birth_rate - death_rate;
```

```python
plot(popt, ':', label='China Population')
    
decorate(xlabel='Year',
         ylabel='Total population (billion)')
```

```python
# Linear Growth Model
# 
# For this model, we'll use the average growth per year from 1960-1979 to simulate the growth of the population if it hadn't 
# been affected by the one-child policy.

# First, we find the average growth per year from 1960-1979, before the one-child policy was enacted.
growth = (popt[t_ocp] - popt[t_0]) / (t_ocp - t_0)

#Establishing the system and its parameters (see previous cells for more information)
system = System(t_0 = t_0,
                t_ocp = t_ocp,
                t_end = t_end,
                p_0 = p_0,
                p_end = p_end,
                growth = growth)

# Now for the update function. We're just going to add the avg growth per year to the year previous:
def update_func_linear(pop, t, system):
    return pop + system.growth;

# Our simulation function runs through each year and then puts our model outputs into a TimeSeries so we can plot it.
def run_simulation(system, update_func):
    results = TimeSeries()
    results[system.t_0] = system.p_0
    
    for t in linrange(system.t_0, system.t_end):
        results[t+1] = update_func(results[t], t, system)
        
    return results

# Lastly, we run and then plot the model.
results = run_simulation(system, update_func_linear)

plot(popt, ':', label='China pop')
plot(results, '--', label='Linear model of pop')
```
