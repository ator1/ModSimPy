```python
R = 0
L = 0
E = 0
V = 0
Gamma = 0
tau = 0
mu = 0
beta = 0
rho = 0
alpha = 0
delta = 0
pi = 0
sigma = 0

def calc_dR(Gamma, tau, mu, beta):
    dR = Gamma * tau - mu * state.R - beta * state.R * state.V
    return dR

def calc_dL(rho, beta, mu, alpha):
    dL = rho*beta*state.R*state.V - mu *state.L - alpha*state.L
    return dL

def calc_dE(rho, beta, alpha, delta):
    dE = (1 - rho)*beta*state.R*state.V + alpha*state.L + delta*state.E
    return dE

def calc_dV(pi, sigma):
    dV = pi*state.E - sigma *state.V
    return dV

state = State(R = R,
             L = L,
             E = E,
             V = V)
```
