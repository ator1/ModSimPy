```python
Gamma = 1.36
tau = 0.2
mu = 0.00136
beta = 0.00027
rho = 0.1
alpha = 0.036
delta = 0.33
pi = 100
sigma = 2
```

```python
def make_system(Gamma, tau, mu, beta, rho, alpha, delta, pi, sigma):
 
    init = State(R=200, L=0, E=0, V=100)
  
    t0 = 0
    t_end = 120
    dt = 0.5

    return System(init=init, t0=t0, t_end=t_end, dt =dt,
                  Gamma = Gamma, tau = tau, mu = mu, beta=beta, rho = rho, alpha = alpha, delta = delta, pi = pi, sigma = sigma)
```

```python
def update_func(state, t, system):
    R, L, E, V = state
    
    unpack(system)
    
    dRdt = (Gamma * tau) - (mu * R) - (beta * R * V)
    dLdt = (rho * beta * R * V) - (mu * L) - (alpha * L)
    dEdt = ((1 - rho) * beta * R * V) + (alpha * L) - (delta * E)
    dVdt = (pi * E) - (sigma * V)
    
    R += dRdt * dt
    L += dLdt * dt
    E += dEdt * dt
    V += dVdt * dt
    
    return State(R=R, L=L, E=E, V=V)
```

```python
system = make_system(Gamma, tau, mu, beta, rho, alpha, delta, pi, sigma)
```

```python
def run_simulation(system, update_func):

    unpack(system)
    
    frame = TimeFrame(columns=init.index)
    frame.row[t0] = init
    
    for t in linrange(t0, t_end):
        frame.row[t+1] = update_func(frame.row[t], t, system)
    
    return frame
```

```python
results = run_simulation(system, update_func)
```

```python
def plot_results(R, L, E):
    plot(R, 'm-', label = "R")
    plot(L, 'g-', label = "L")
    plot(E, 'b-', label = "E")
    decorate(title = "HIV Infection",
            xlabel = "Time",
            ylabel = "Number of Cells")
```

```python
plot_results(results.R, results.L, results.E)
```
